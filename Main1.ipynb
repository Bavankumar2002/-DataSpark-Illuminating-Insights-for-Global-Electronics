```python
table1 = '''
CREATE TABLE Customers_table (
    CustomerKey INT,
    Gender VARCHAR(150),
    Name VARCHAR(150),
    City VARCHAR(150),
    StateCode VARCHAR(150),  
    State VARCHAR(150),
    ZipCode INT,
    Country VARCHAR(150),
    Continent VARCHAR(150),
    Birthday DATE 
)
'''
cursor.execute(table1)

```

```python
alter_table = "ALTER TABLE Customers_table MODIFY ZipCode VARCHAR(10);"
cursor.execute(alter_table)

# Assuming cus_df is already defined
# Rename the columns if needed
cus_df.rename(columns={
    'State Code': 'StateCode',  # Rename to remove spaces
    'Zip Code': 'ZipCode'        # Rename to remove spaces
}, inplace=True)

# Corrected INSERT statement
data1 = '''
INSERT INTO Customers_table (
    CustomerKey, Gender, Name, City, StateCode, State, ZipCode, Country, Continent, Birthday
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
'''

# Loop to insert each row from the DataFrame
for index, row in cus_df.iterrows():
    cursor.execute(data1, (
        row['CustomerKey'],
        row['Gender'],
        row['Name'],
        row['City'],
        row['StateCode'],
        row['State'],
        row['ZipCode'],    # Now will accept alphanumeric postal codes
        row['Country'],
        row['Continent'],
        row['Birthday']
    ))

# Commit the transaction
con.commit()
```
